## K-Means algorithm [18 marks]

### Implement the functions [5 marks]

For this, you will only use functions from the NumPy library. You will implement four functions to realize the K-Means algorithm.

- `distanceFunc`: This function accepts your data and cluster centers and returns the squared pairwise distances between data points and cluster centers. 
- `KMinit`: This function accepts your data and returns the initial cluster centres. You will use the "greedy approach" described on page 6-16 in the textbook, but instead of a randomly selected data point always pick the first entry in the data set as the first center. 
- `lossFunc`: This function accepts the squared pairwise distances between data points and cluster centers and returns the sum of cluster errors as defined in (6.5) in the textbook. 
- `KMmeans`: This function implements the K-Means algorithm as described in the textbook. It returns the cluster centers and the sum of cluster errors as defined in (6.5) in the textbook. 

```python
def distanceFunc(x, mu):
    # Inputs  
    # x: is an NxD data matrix (N observations and D dimensions)
    # mu: is an KxD cluster center matrix (K cluster centers and D dimensions)
    # Output
    # pair_dist2: is the NxK matrix of squared pairwise distances
    
    # YOUR CODE HERE
    # Initialize pair_dist2
    pair_dist2 = np.zeros([x.shape[0],mu.shape[0]])
    
    for i in range(0,mu.shape[0]):
        pair_dist2[:,i] = np.sum(np.square(x-mu[i,:]),axis=1)
    
    return pair_dist2
```

```python
def KMinit(x, K):
    # Inputs
    # x: is an NxD data matrix 
    # K: number of clusters
    # Output
    # mu: is the KxD matrix of initial cluster centers using the "greedy approach" described on page 6-16 in the textbook. 
    # Remark: Always pick the first entry in the data set as the first center. 
    
    # YOUR CODE HERE
    # Initialize mu
    mu = np.zeros([K,x.shape[1]])
    
    for i in range(0,K):
        if i==0:
            mu[i,:] = x[i,:]
            centroid = x[i,:]
            continue
        maxD = 0
        maxDind = 0
        
        for j in range(0,x.shape[0]):
            dist = np.linalg.norm((centroid-x[j,:]),ord=2)
            if dist > maxD:
                # Check if Point is Already in Mu
                if(np.isin(x[j,:],mu).all()):
                    continue
                maxD = dist
                maxDind = j

        # Set Mu
        mu[i,:] = x[maxDind,:]
        
        tempCent = np.zeros([1,x.shape[1]])
        for k in range(0,i+1):
            tempCent = tempCent + mu[k,:]
        centroid = tempCent / (i+1)
    
    return mu
```

```python
def lossFunc(pair_dist2):
    # Input 
    # pair_dist2: is an NxK matrix of squared pairwise distances
    # Output
    # loss: error as defined in (6.5) in the textbook
    
    # YOUR CODE HERE
    # Initialize Loss
    loss = 0
    # Column Sum to Get Error per Cluster
    loss = np.sum(np.min(pair_dist2,axis=1))
    return loss
```

```python
def Kmeans(x,K):
    # Inputs
    # x: is an NxD data matrix 
    # K: number of clusters
    # Outputs
    # mu: is the KxD of cluster centers  
    # loss: error as defined in (6.5) in the textbook 
    
    # YOUR CODE HERE
    # Initialize Loss
    loss = []
    
    # 1. Generate Initial Cluster Centres
    mu = KMinit(x,K)
    
    iter = 0
    while(1):
        iter = iter + 1
        # 2. Create Clusters
        pair_dist2 = distanceFunc(x,mu) # Compute Pair-wise Distance
        clusterIndex = (np.argmin(pair_dist2,axis=1)) # Map Data to Clusters
        # Clusters can be explicitly defined as x[clusterIndex==n,:]
        loss.append(lossFunc(pair_dist2))
        
        if(iter>1):
            if loss[iter-1]==loss[iter-2]:
                break
    
        # 3. Update Cluster Center to Centroid of Cluster
        for i in range(0,mu.shape[0]):
            mu[i,:] = np.mean(x[clusterIndex==i,:],axis=0)
        
    # Repeat 2 and 3 Until Ein stops decreasing
    
    return mu, loss  
```

### Run K-Means on 2D data set [4]

Implement a script that runs the K-Means algorithm (using the functions above) with $K = 1, 2, 3, 4, 5$ for the data set data2D.npy.  

The script should produce the following plots:
- for each of these values of $K$, a 2D scatter plot of the data points colored by their cluster assignments plus the cluster centers
- the K-means loss as a function of $K$

```python
# Helper Function To Plot
def plotClust(x,mu,loss):
    pair_dist2 = distanceFunc(x,mu)
    clusterIndex = (np.argmin(pair_dist2,axis=1))
    
    fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(15, 6))
    
    color=iter(plt.cm.rainbow(np.linspace(0,1,mu.shape[0])))
    for i in range(0,mu.shape[0]):
        cluster = x[clusterIndex==i,:]
        ax1.scatter(cluster[:,0],cluster[:,1],next(color),label="Cluster-{}".format(i+1))
    ax1.scatter(mu[:,0],mu[:,1],color="black",marker="x",label="Cluster Center")
    ax1.set_xlabel('X1')
    ax1.set_ylabel('X2')
    ax1.legend(loc="lower right")
    ax1.set_title("K = {}".format(mu.shape[0]))
    
    ax2.plot(loss)
    ax2.set_xlabel('Iterations')
    ax2.set_ylabel('K-Means Loss')
    ax2.set_title("K-Means Loss vs Number of Iterations | K = {}".format(mu.shape[0]))

# Import Data
data2D = np.load('data2D.npy')

finalLoss = np.zeros([5,1])
for i in range(1,6):
    mu, Loss = Kmeans(data2D,i)
    finalLoss[i-1] = Loss[-1]
    plotClust(data2D,mu,Loss)

# Loss vs K
plt.figure()
plt.plot(np.linspace(1,5,5),finalLoss,marker='o')
plt.xlabel("K")
plt.xticks(np.linspace(1,5,5))
plt.ylabel("K-Means Loss")
plt.title("K-Means Loss as a Function of K")
plt.show()
```

Discuss below on how many clusters you think is "best" and why? 

**Solution:**

We can see from the final plot, that the Sum of Squared Error loss or K-Means Loss keeps decreasing as K increases. The loss will eventually be equal to zero once there are as many clusters as there are data points.

One way of making an estimate about the true number of clusters is by using the "elbow" method. The "elbow" here refers to a sharp and usually noticeable change in the graph beyond which the graph starts to fall gradually.

The "elbow" method is not really a reliable indicator since for most real world data, like the dataset we are testing with, an elbow is not easily discernable. Different individuals may also have different notions about where the elbow may exactly lie for the same graph.

In the graph above there is a sharp decrease in the K-Means loss from K = 1 to K = 2. This could lead one to declare that K = 2 is the so-called "best" estimate for the number of clusters for the dataset. On closer inspection, one can see that the decrease in K-Means loss from K = 2 to K = 3 is also very high relative to subsequent decreases. So one could also say that K = 3 could be a fairly plausible estimate. The elbow method is not really well-defined and cannot help reconcile this kind of ambiguity.

For the given dataset, I would be inclined to pick K = 2 since the elbow is more pronounced at K = 2. That said, K could just as well be 3.

```python
# Note 
# The section of the code computing the gap statistic has been amended based on the update shared by Dr.Lampe on 21-March-2020 
# per his adivice, "please first take the logarithm for each error value and then average."

np.random.seed(421)

# Gap Statistics

# Init Result Vector
resErr = np.zeros([10,5])

# Determine Bounding Box Around Data2D
def boxBound(x):
    x1min = np.min(data2D[:,0])
    x1max = np.max(data2D[:,0])
    x2min = np.min(data2D[:,1])
    x2max = np.max(data2D[:,1])
    return x1min, x1max, x2min, x2max

x1min, x1max, x2min, x2max = boxBound(data2D)

# Loop Over 10 Datasets
for i in range(0,10):
    
    # Init Dataset
    dataRandX = np.random.uniform(low=x1min,high=x1max,size=(data2D.shape[0],1))
    dataRandY = np.random.uniform(low=x2min,high=x2max,size=(data2D.shape[0],1))
    dataRand2D = np.concatenate((dataRandX,dataRandY),axis=1) 
    
    # Loop Over K = 1 to 5
    for j in range(1,6):
        muRand, LossRand = Kmeans(dataRand2D,j)
        resErr[i,j-1] = LossRand[-1]

# Find Average of K-Means Error Per Cluster Size
avgKmeans = (np.mean(resErr,axis=0,keepdims=True)).T



# Compute Gap Statistic
def gapStat(randKmeansError,KmeansError):
    # Assume randKmeansError is already average of Logs
    res = (randKmeansError) - np.log(KmeansError)
    return res

# Update for Computing Gap Stats (Mean of Logs Computed instead of Log of Mean)
newResErr = np.log(resErr)
newAvgKmeans = (np.mean(newResErr,axis=0,keepdims=True)).T
gap = gapStat(newAvgKmeans,finalLoss)

# Plot Graphs
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(17, 6))

# Plot Avg K-Means Error
ax1.plot(finalLoss,marker='o',label='$E_{in}(K)$')
ax1.plot(avgKmeans,marker='o',label='$E_{in}^{Rand}(K)$')
ax1.set_xticks(np.arange(5), minor=False)
ax1.set_xticklabels(np.linspace(1,5,5), fontdict=None, minor=False)
ax1.set_xlabel("K")
ax1.set_ylabel("K-Means Error")
ax1.legend(loc="upper right")
titax1 = ax1.set_title("K-Means Error as a Function of K")

# Plot Gap Statistics
ax2.plot(gap,marker='o')
ax2.set_xticks(np.arange(5), minor=False)
ax2.set_xticklabels(np.linspace(1,5,5), fontdict=None, minor=False)
ax2.set_xlabel("K")
ax2.set_ylabel("Gap Statistic")
titax2 = ax2.set_title("Gap Statistic as a Function of K")
```

Briefly explain the "gap statistic" and discuss on how many clusters you think is "best" based on the gap statisic and on how this compares with your result from above.  

**Solution:**

The gap statistic is a formal and statistical way of estimating the number of clusters within a given dataset. The rationale is to standardize the graph of $\log E_{in}(K)$ (where $E_{in}(K)$ is the K-Means loss) by comparing it with its expectation under a null reference distribution of data (data that has no obvious clustering). The estimate for the optimal number of clusters is the value of K for which $\log E_{in}(K)$ falls the farthest below the curve for the reference data. The gap statistic is defined as,

$$G(K)= Expectation \left( \log \left[E_{in}^{rand(i)}(K)\right]\right)-\log \left[E_{in}(K)\right]$$

where $E_{in}^{rand(i)}(K)$ denotes the K-Means loss of the i-th reference dataset. 

Multiple reference datasets are created by drawing data randomly from a uniform distribution that is confined by the smallest bounding box enclosing the dataset being analysed.

The optimal value of K is the value of K such that maximizes the gap statistic.

For the given 'data2D' dataset, the gap statistic is maximum at K = 3. Therefore this is the best estimate for the number of clusters. While the elbow method could not be used to make a conclusive estimate between K = 2 and K = 3, the gap statistic indicates that K = 3 is the optimal number of clusters for the given data.

First, write a script that, for the 2D data set data2D.npy,
- estimates the parameters of the Gaussian mixture models with $K=3$ components,
- plots the contours of the density,
- plots the decision boundaries for clustering. 

In the markdown cell below, interpret what you see in the plot and compare with the plot for $K=3$ clusters for the K-Means algorithm above. 

```python
# Import GMM
from sklearn.mixture import GaussianMixture
from matplotlib.colors import LogNorm, Normalize # For Plotting Contour Lines
import matplotlib as mpl # For Plotting Decision Boundary

# Define Function to Return Fitted GMM for Given Data and K
def gaussMix(data,K):
    # Create GMM
    gmm = GaussianMixture(n_components=K)
    gmm.fit(data)
    
    return gmm

# Get GMM on data2D for K = 3
K = 3
gmm = gaussMix(data2D,K)

# Plotting Contours of Density
X, Y = np.meshgrid(np.linspace(-4,5),np.linspace(-6,3))
XY = np.array([X.ravel(), Y.ravel()]).T
Z = -gmm.score_samples(XY)
Z = Z.reshape(X.shape)
# Visualize

fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(17, 6))
cont = ax1.contour(X,Y,Z,norm=Normalize(vmin=0, vmax=30),levels=np.logspace(0, 1.47, 15))
ax1.scatter(data2D[:,0],data2D[:,1], s=1)
ax1.set_title("Contours of Density")
ax1.set_xlabel('X1')
ax1.set_ylabel('X2')

ax2.scatter(data2D[:,0],data2D[:,1], c = gmm.predict(data2D), s=1)
ax2.set_xlabel('X1')
ax2.set_ylabel('X2')
ax2.set_title('Decision Regions for Clustering (95% Confidence)')
colors = ['red', 'turquoise', 'darkorange']
for n, color in enumerate(colors):
    covariances = gmm.covariances_[n][:2, :2]
    v, w = np.linalg.eigh(covariances)
    u = w[0] / np.linalg.norm(w[0])
    angle = np.arctan2(u[1], u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    v = 2. * np.sqrt(2.) * np.sqrt(v)
    ell = mpl.patches.Ellipse(gmm.means_[n, :2], v[0], v[1],180 + angle, color=color)
    ell.set_clip_box(ax2.bbox)
    ell.set_alpha(0.2)
    ax2.add_artist(ell)
    ax2.set_aspect('equal', 'datalim')
plt.show()

# Decision Regions
X2, Y2 = np.meshgrid(np.linspace(-4,5,1000),np.linspace(-6,3,1000))
XY2 = np.array([X2.ravel(), Y2.ravel()]).T
Z2 = gmm.predict(XY2)
Z2 = Z2.reshape(X2.shape)

plt.figure(figsize=(17, 6))
plt.imshow(Z2,origin='lower')
plt.title('Decision Regions Colored Over a Fine Meshgrid')
plt.xticks([])
plt.yticks([])
plt.locator_params(axis='both', nbins=10)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

**Solution:**

Gaussian Mixture Models (GMMs) are probabilstic models which makes an assumption that all data points are generated a mixture of finite Gaussian distributions whose parameters (mean and standard deviation) are unknown. For K = 3, we assume for their to be 3 finite gaussian distributions and their parameters are estimated by using the gmm.fit() function.

The contour plot above is a representation of the probability density functions of each of the 3 gaussian distributions for the GMM. The radius of the concentric rings increase logarithmically as one goes away from the center. The color of the rings indicate the -log(probability) at that point. The rings are darker when close to the center (indicating a higher probability) and become lighter as one moves away from the center (indicating a lower probability). For purposes of illustration, 15 rings have been plotted for each distribution.

The second plot displays the dataset colored based on cluster assignments. Additionally, a colored transparent disk has been overlaid to indicate the region within which one can determine with 95% certainty that a point belongs to that cluster.

The third plot paints decision regions based on the estimated GMM. A fine meshgrid of points were first generated and then the GMM was used to predict the cluster assignment of every point with the meshgrid.

Compared to the result of KMeans clustering for K = 3, the GMM solution seems more natural and likely. This could be because GMMs are a generalization of KMeans that pays attention to the covariance structure of the data.

Write a script that trains the Gaussian mixture model for the data set data2D.npy for $K=1,2,3,4,5$, and computes and plots the Bayesian Information Criterion (BIC) as a function of $K$.

In the markdown cell below, explain the BIC and what model size it suggests, and compare with your result for the K-Means algorithm above. 

```python
# Loop over K = 1 to 5 and Plot BIC vs K
finBIC = np.zeros([5,1])

for i in range(1,6):
    gmmFin = gaussMix(data2D,i)
    finBIC[i-1,0] = gmmFin.bic(data2D)

# Plot BIC vs K
plt.figure()
plt.plot(finBIC,marker='o')
plt.xticks(np.arange(5),np.linspace(1,5,5))
plt.xlabel('K')
plt.ylabel('BIC')
plt.title('Bayesian Information Criterion (BIC) as a Function of K')
plt.show()
```

**Solution:**

The Bayesian Information Criterion (BIC) is a criterion for model selection which is based on the maximum likelihood of the models being analyzed. The goal is to choose the model which has the lowest BIC score.

As can be seen from the plot above, the lowest BIC is for the model with K = 3. Therefore, K = 3 is the best estimate for the number of clusters.

The results of both K Means and GMM are similar in that both approaches estimate that the given dataset can be clustered into 3 groups. They vary slightly with respect to the cluster assignment of each datapoint.
